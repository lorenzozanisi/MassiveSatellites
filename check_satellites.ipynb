# initialising at accretion

```python
make = make_satellites(use_acc=True)
```

```python
mu = 2.
df_high = make.make_censat(Mstar_low=11.5, Mstar_up=12.5, mu=mu, AK=0.012, AK_SF=0.015)
```

### high mass

```python
fig,ax = plt.subplots(1,1)
ax = plot_sizefunctions(df_high, ax, m=11.75,TType='All')

fig.tight_layout()
plt.title(r'$\log{M_{\rm star}}/M_\odot>11.5$')
plt.text(1.8,4.e-7,'satellites\ninitialized\nat accretion')
plt.savefig('Pictures/Sizefunct_11.75_Macc_all.png')
```

```python
fig,(ax1,ax2) = plt.subplots(1,2, figsize=(32,16), sharey=True)
ax1 = plot_sizefunctions(df_high, ax1, m=11.75,TType='LTGs')
ax2 = plot_sizefunctions(df_high, ax2, m=11.75,TType='ETGs')

fig.tight_layout()
ax2.set_ylabel('')
ax1.set_title('LTGs')
ax2.set_title('ETGs')
fig.suptitle(r'$\log{M_{\rm star}}/M_\odot>11.5$ '+r'$\mu=$'+str(mu),y=1.1, )
#ax1.text(1.8,4.e-7,'satellites\ninitialized\nat accretion')
plt.savefig('Pictures/Sizefunct_11.75_Macc_fquench.png')
```

### low mass

```python
mu = 2.3
df_low = make.make_censat(Mstar_low=11.2, Mstar_up=11.5, mu=mu, AK=0.01, AK_SF=0.015, sigmaK=0.09, sigmaK_SF=0.011)
```
